### Encoding categorical data

Categorical variables/features are any feature type can be classified into two major types:
- Nominal
- Ordinal

Nominal variables are variables that have two or more categories which do not have any kind of order associated with them. For example, if gender is classified into two groups, i.e. male and female, it can be considered as a nominal variable.

```python
X = df.drop('Exited', axis=1)
y = df['Exited']
```

Label Encoding the "Gender" column

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
X['Gender'] = le.fit_transform(X['Gender'])
```

One Hot Encoding the "Geography" column

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), ['Geography'])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

### Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Part 2 - Training the different model and comparing their accuracy scores

The focus of this project is to look at how the different models performs best on our data. We will compare the performance of the Decision Trees, Random Forests, SVM Models and the Artificial Neural Network.

## Part 3 - Training the ANN

### Compiling the ANN

```python
pip install scikeras
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from scikeras.wrappers import KerasClassifier

# Define model-building function that accepts kwargs
def build_ann(units=6):  # can add more hyperparameters
    model = Sequential()
    model.add(Input(shape=(X_train.shape[1],)))
    model.add(Dense(units=units, activation='relu'))
    model.add(Dense(units=units, activation='relu'))
    model.add(Dense(units=1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

```

# KerasClassifier from scikeras

```python

ann = KerasClassifier(model=build_ann, epochs=100, batch_size=32, verbose=0, random_state=0)

# Fit model
ann.fit(X_train, y_train)
```

### Training the ANN on the Training set

## Part 4 - Making the predictions and evaluating the model

### Predicting the result of a single observation

**Homework**

Use our ANN model to predict if the customer with the following informations will leave the bank:

Geography: France

Credit Score: 600

Gender: Male

Age: 40 years old

Tenure: 3 years

Balance: \$ 60000

Number of Products: 2

Does this customer have a credit card? Yes

Is this customer an Active Member: Yes

Estimated Salary: \$ 50000

So, should we say goodbye to that customer?

**Solution**

```python
print(ann.predict(sc.transform([[1, 0, 0, 600, 1, 40, 3, 60000, 2, 1, 1, 50000]])) > 0.5)
```

Therefore, our ANN model predicts that this customer stays in the bank!

**Important note 1:** Notice that the values of the features were all input in a double pair of square brackets. That's because the "predict" method always expects a 2D array as the format of its inputs. And putting our values into a double pair of square brackets makes the input exactly a 2D array.

**Important note 2:** Notice also that the "France" country was not input as a string in the last column but as "1, 0, 0" in the first three columns. That's because of course the predict method expects the one-hot-encoded values of the state, and as we see in the first row of the matrix of features X, "France" was encoded as "1, 0, 0". And be careful to include these values in the first three columns, because the dummy variables are always created in the first columns.

### Predicting the Test set results

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(-1, 1), y_test.to_numpy().reshape(-1, 1)), axis=1))
```

### Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

## Applying k-Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = ann, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

## Applying Grid Search to find the best model and the best parameters

```python
from sklearn.model_selection import GridSearchCV
parameters = {
    'model__units': [6, 12],          # number of neurons per layer
    'batch_size': [16, 32],           # batch sizes
    'epochs': [50, 100]               # training epochs
}

# Step 4: Set up GridSearchCV
grid_search = GridSearchCV(estimator=ann,
                           param_grid=parameters,
                           scoring='accuracy',
                           cv=5,
                           n_jobs=-1)
-
# Step 5: Fit model
grid_search.fit(X_train, y_train)

# Step 6: Print best results
best_accuracy = grid_search.best_score_
best_parameters = grid_search.best_params_

print("Best Accuracy: {:.2f} %".format(best_accuracy * 100))
print("Best Parameters:", best_parameters)
```

## Using Decission tree Classifier
A decision tree is a supervised learning algorithm that is perfect for classification problems, as its able to order classes on a precise level. It works like a flow chart, separating data points into two similar categories at a time from the tree trunk to branches, to leaves, where the categories become more finitely similar. This creates categories within categories, allowing for organic classification with limited human supervision.

## Training the Decision Tree Classification model on the Training set
